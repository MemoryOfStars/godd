Add docking ligand pdb files to protein pdb files

```python
import os
```

```python
ligand_path = '/home/kmk_gmx/Desktop/bioinfo/negative_ligand_pdb/'
protein_path = '/home/kmk_gmx/Desktop/bioinfo/refined-set/'
tar_path = '/home/kmk_gmx/Desktop/bioinfo/negative_pdb/'
for ligand_fname in os.listdir(ligand_path):
    ligand_name = ligand_fname[:-4]
    ligand_pdb = []
    pdbid = ligand_name[:4]
    with open(ligand_path+ligand_fname,'r') as ligand_file:
        ligand_pdb = ligand_file.readlines()
    protein_pdb = []
    with open(protein_path+pdbid+'/'+pdbid+'_protein.pdb','r') as protein_file:
        protein_pdb = protein_file.readlines()
    tar_file_content = []
    for line in protein_pdb:
        if (line[:4] == 'ATOM' or (line[:6]=='HETATM' and line[17:20]!='HOH')) and line[21:22]=='A' and line[76:78].strip() != 'H':
            tar_file_content.append(line)
    for line in ligand_pdb:
        if (line[:4] == 'ATOM' or (line[:6]=='HETATM' and line[17:20]!='HOH')) and line[21:22]=='A' and line[76:78].strip() != 'H':
            tar_file_content.append(line)
    with open(tar_path+ligand_fname,'w+') as tar_file:
        tar_file.writelines(tar_file_content)
    print(ligand_name)
```
